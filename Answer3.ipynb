```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Phasors

```python
intensity_list = []
amplitude_list = []


# CREATING 19200 PHASORS
for i in range(19200):
    real, imaginary = 0, 0
    
    # ADDING 10 PHASORS WHERE AMPLITUDE AND THETA ARE TAKEN RANDOMLY USING A UNIFORM DISTRIBUTION
    for j in range(10):
        amplitude = np.random.uniform(0, 10) 
        theta = np.random.uniform(0, 5 * np.pi)
        
        # PHASORS CAN BE DEFINED BY COMPLEX NUMBERS 
        # HENCE WE CAN CALCULATE THEN USING REAL (cos) AND IMAGINARY (sin)
        real += amplitude * np.cos(theta)
        imaginary += amplitude * np.sin(theta)
    
    
    # ADDING THE SQUARE OF THE REAL AND IMAGINARY AND STORING IT IN THE INTENSITY LIST
    intensity = np.sum([pow(real, 2), pow(imaginary, 2)])
    intensity_list.append(intensity)
    
    
    # COMPUTING THE SQUARE ROOT OF THE INTENSITY AND APPENDING IT TO THE AMPLITUDE
    amplitude_list.append(math.sqrt(intensity))
```

```python
# PLOTTING THE INTENSITY VALUES ON A HISTOGRAM
fig, ax = plt.subplots()
plt.title('Intensity distribution')
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
ax.hist(intensity_list, bins=100)
plt.xlim(0, 2200)
plt.ylim(0, 2200)
plt.show()
```

```python
# PLOTTING THE AMPLITUDE VALUES ON A HISTOGRAM (Rayleigh distribution)
fig, ax = plt.subplots()
plt.title('Rayleigh distribution')
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
ax.hist(amplitude_list, bins=100)
plt.xlim(0, 60)
plt.ylim(0, 550)
plt.show()
```

```python
# DISPLAYING THE IMAGE OF THE INTENSITY VALUES
fig, ax = plt.subplots()
plt.title('Intensity distribution')
ax.set_xticks([])
ax.set_yticks([])
intensity_list = np.array(intensity_list).reshape(160, 120)
plt.imshow(intensity_list, cmap='gray')
plt.show();
```

```python
# DISPLAYING THE IMAGE OF THE AMPLITUDE VALUES
fig, ax = plt.subplots()
plt.title('Rayleigh distribution')
ax.set_xticks([])
ax.set_yticks([])
amplitude_list = np.array(amplitude_list).reshape(160, 120)
plt.imshow(amplitude_list, cmap='gray')
plt.show();
```
