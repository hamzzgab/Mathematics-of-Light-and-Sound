```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft2
```

# Creating an Image

- A triangle image is created using Paint

```python
# CREATING AN IMAGE SHIFT 
# DISPLAYING HIGH FREQUENCIES IN THE CENTER OF THE TRANSFORMED IMAGE
def FFT_SHIFT(img):
    for i in range(0, img.shape[0]):
        for j in range(0, img.shape[1]):
            img[i][j] = img[i][j] * pow(-1, (i + j))
    return img
```

```python
# READIG THE IMAGE USING MATPLOTLIB
img = plt.imread("/Users/hamzz/Desktop/AVR/Mathematics "
                 "of Light and Sound/Exam/triangle2.png")

# USING JUST THE FIRST CHANNEL OF THE IMAGE 
# AND PASSING IT TO THE FUNCTION CREATED ABOVE
# CONVERTING IT TO ABSOLUTE VALUES AS IT 
# CONSISTS OF NEGATIVE INTEGER VALUES
img = FFT_SHIFT(img[:, :, 0])
plt.imshow(np.abs(img), cmap='gray')
plt.show()
```

```python
# USING SCIPYS FFT2 FOR FOURIER TRANSFORM
img = fft2(img)
plt.imshow(np.abs(img), cmap='gray')
plt.show()
```
