# Intensity Sum
I have used the following formula for calculating the intensity sum with two conditions 
1. $\lambda_x != \lambda_y$
1. $\lambda_x = \lambda_y$

$$\large f_z(z)=   \left\{ \begin{array}{ll}{\lambda_x \lambda_y \over \lambda_y-\lambda_x} (e^{-\lambda_x z} - e^{-\lambda_y z})&\lambda_y\neq \lambda_x \\z\lambda^2e^{\lambda z}  & \lambda = \lambda_x = \lambda_y\\\end{array} 
\right.  $$

```python
# WHEN lambdax != lambday

fig, ax = plt.subplots()

lambdax = 1
lambday = 2

# CREATING A LINSPACE FOR Z IN THE FORMULA
zi = np.arange(0, 10, 0.1)

x = np.exp(-zi * lambdax) * lambdax
y = np.exp(-zi * lambday) * lambday


# CALCULATING THE INTENSITY SUM WHEN LAMBDA VALUES ARE DIFFERENT
z = ( (lambdax * lambday) / (lambday - lambdax)) * (np.exp(-lambdax * zi)  - np.exp(-lambday * zi))

ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
plt.title('$\lambda_x != \lambda_y$')
ax.plot(zi, x, label="I1", lw=1, linestyle="--", color='b')
ax.plot(zi, y, label="I2", lw=1, linestyle="--", color='r')
ax.plot(zi, z, label="I1+I2", lw=2, linestyle="-", color="green")
plt.legend()
plt.show()
```

```python
# WHEN lambdax = lamday
fig, ax = plt.subplots()

_lambda = 1

# CREATING SIMILAR LAMBDAS
x, y = np.exp(-zi * _lambda) * _lambda, np.exp(-zi * _lambda) * _lambda

# CALCULATING THE INTENSITY SUM WHEN LAMBDA VALUES ARE SAME
z = zi * _lambda ** 2 * np.exp(-zi * _lambda)

ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
plt.title('$\lambda = \lambda_x = \lambda_y$')
ax.plot(zi, x, label="I1&I2", lw=1, linestyle="--", color="r")
ax.plot(zi, z, label="I1+I2", lw=2, linestyle="-", color="g")
plt.legend()
plt.show()
```

# Plotting the Normal Distribution
- The code then creates two variables, x and y, which are both set to zero with a size of 100000.
- Next, we iterate through each number in our list of integers from 0-100000.
- For each iteration through this loop, we add one random number between -10 and 10 into our X variable using np.random(), and do the same thing for our Y variable using np.random().
    - These variables represent random numbers between -10 and 10 with uniform distribution.

```python
fig = plt.figure();

x = np.zeros(100000)
y = np.zeros(100000)
for i in range(0, 100000):
    
    # ASSIGNING RANDOM VALUES UNIFORMLY FOR 100000 VALUES
    x[i] = np.random.uniform(-10, 10)
    y[i] = np.random.uniform(-10, 10)


# CREATING A NEW LIST OF ZEROS OF LENGTH 10000
X = np.zeros(10000)
Y = np.zeros(10000)

# LOOPING OVER THE NEW LIST
for i in range(0, 10000):
    # STORING THE SUM OF 10 VALUES IN THE NEWLY CREATED LIST
    for j in range(0, 10):
        X[i] += x[10*i + j]
        Y[i] += y[10*i + j]
      
    
# SETTING THE AXES
ax = plt.gca()
ax.set_aspect(1)
plt.xlim(-100, 100)
plt.ylim(-100, 100)
plt.scatter(X, Y, s=0.1, c='black')
plt.show()
```
