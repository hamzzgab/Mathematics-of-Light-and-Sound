```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
```

# Creating Huygens-Fresnel Construction

- To create this plot, I have created an array of radius's and for each of these radius's, I have created a semi-circle and plotted them on their respective x and y coordinate

```python
theta = np.arange(0, np.pi, 0.01)
X = np.linspace(0, 5, 6)
ax = plt.gca()
plt.title('Huygens-Fresnel Construction')


# SETTING THE AXES OF THE PLOT
ax.spines["right"].set_color("none")
ax.spines["left"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.spines["bottom"].set_position(("data", -3))
ax.set_aspect(1)


# ADDING THE BLUE RECTANGLE FROM WHICH THE WAVE WOULD START
ax.add_patch(patches.Rectangle(xy=(-3, 0), width=11, height=0.08, color="blue"))


# DIFFERENT RADIUS VALUES
radius = [-1/2, -1, -3/2, -2, -5/2, -3]


# CREATING THE STARTING WAVE
for r in radius:
    for i in X:
        cx = X[int(i)]+r*np.cos(theta)
        cy = r * np.sin(theta)
        plt.plot(cx, cy, 'grey')
        
        
# MARKER POINTS INDICATING THE CENTER OF THE WAVE
ax.plot(X,[0]*6,'y.', markersize=14)


# CREATING THE DIFFRACTED WAVE
for r in radius:
    for i in X:
        cx = X[int(i)]+r*np.cos(theta)
        cy = r * np.sin(theta)
        plt.plot(cx, cy - 3, 'grey')

        
# MARKER POINTS INDICATING THE CENTER OF THE WAVE
ax.plot(X,[-3]*6,'y.', markersize=14)


# ADDING THE BLUE RECTANGLE AT WHICH THE WAVE ENDS
ax.add_patch(patches.Rectangle(xy=(-3, -6), width=11, height=0.08, color="blue"))
plt.show();
```

```python
theta = np.arange(0, np.pi, 0.01)
X = np.linspace(0, 5, 6)
ax = plt.gca()
plt.title('Diffraction')


# SETTING THE AXES OF THE PLOT
ax.spines["right"].set_color("none")
ax.spines["left"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.spines["bottom"].set_position(("data", 0))
ax.set_aspect(1)


# SETTING THE LINES OF THE WAVE
x = np.linspace(-5,10,100)
y1 = np.linspace(0.5,0.5,100)
y2 = np.linspace(1,1,100)
y3 = np.linspace(1.5,1.5,100)
plt.plot(x, y1, color="blue", linewidth=1, alpha=0.7)
plt.plot(x, y2, color="blue", linewidth=1, alpha=0.7)
plt.plot(x, y3, color="blue", linewidth=1, alpha=0.7)


# ADDING THE BLUE RECTANGLE FROM WHICH THE WAVE WOULD START DIFFRACTING
ax.add_patch(patches.Rectangle(xy=(-5, 0), width=4.5, height=0.2, color="grey"))
ax.add_patch(patches.Rectangle(xy=(5.5, 0), width=4.5, height=0.2, color="grey"))


# ARROW TO SHOW TO SHOW WHICH DIRECTION THE WAVE TRAVELS
plt.arrow(2.5, 3.5, 0, -3, color="r", width = 0.1, linestyle='-')


# DIFFERENT RADIUS VALUES
radius = [-1/2, -1, -3/2, -2, -5/2, -3]


# CREATING THE DIFFRACTED WAVE
for r in radius:
    for i in X:
        cx = X[int(i)]+r*np.cos(theta)
        cy = r * np.sin(theta)
        plt.plot(cx, cy, 'grey')

        
# MARKER POINTS INDICATING THE CENTER OF THE WAVE
ax.plot(X,[0]*6,'y.', markersize=14)
plt.show();
```
